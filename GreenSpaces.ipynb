```python
import pandas as pd

df = pd.read_csv("dataSets/GreenSpaces.csv")

# Normalize column names
df.columns = df.columns.str.strip().str.lower()

# Normalize neighborhood names
df['neighborhood'] = df['neighborhood'].astype(str).str.strip().str.lower()

# Get neighborhood and total acreage from dataset
acreage_data = df[['neighborhood', 'total acreage']]

# Total acreage for each neighborhood (sum acreage if multiple rows exist)
acreage_combined = acreage_data.groupby("neighborhood", as_index=False).agg({"total acreage": "sum"})

# Sort by total acreage (smallest to largest)
acreage_sorted = acreage_combined.sort_values(by="total acreage", ascending=True)

# Step 7: Display full output
pd.set_option("display.width", 200)
pd.set_option("display.max_rows", None)

print(acreage_sorted)

```
