```python
import pandas as pd

fireInc = pd.read_csv("FireIncidents.csv")

# Step 3: Normalize column names
fireInc.columns = fireInc.columns.str.strip().str.lower()

# Step 4: Group by neighborhood and count incidents
fire_counts = fireInc.groupby("neighborhood").size().reset_index(name="incident_count")

# Step 5: Sort from least to greatest
fire_counts_sorted = fire_counts.sort_values(by="incident_count", ascending=True)

print(fire_counts_sorted)

```
