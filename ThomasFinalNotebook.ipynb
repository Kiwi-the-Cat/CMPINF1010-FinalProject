my metric was the number of fire incidents in each neighborhood.
my dataset: https://data.wprdc.org/dataset/fire-incidents-in-city-of-pittsburgh/resource/8d76ac6b-5ae8-4428-82a4-043130d17b02 

```python
import pandas as pd
df = pd.read_csv("Fire.csv")

#makes new dataframe with only neighborhood column
neighborhoods = df['neighborhood'].value_counts().tail(20)

#reverses entry order to display from least to greatest
neighborhoods = neighborhoods.iloc[::-1]

#plots graph
neighborhoods.plot.bar(xlabel = 'Neighborhoods',ylabel = 'Fire Incidents',title = "Top 20 neighborhoods with the least fire incidents")
```

Conclusion: This graph displays the number of fire incidents per neighborhood. The graph only shows the top 20 neighborhoods with the least amount of fire incidents. Since my metric for safety was number of fire incidents, we can see that Mount Oliver Borough is the safest neighborhood using this metric as it has the least amount of fire incidents.
