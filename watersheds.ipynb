```python
# Plot Pittsburgh watersheds. Replace `path` with your local file or URL (GeoJSON/Shapefile/etc).
import geopandas as gpd
import matplotlib.pyplot as plt

try:
    import contextily as ctx
    has_ctx = True
except Exception:
    has_ctx = False

path = "/home/kass/Downloads/Programming/Python/1010/CMPINF1010-FinalProject/d6ba2cc7-af46-41ab-9a2e-f8711bfeeabe.csv"  # <-- change to your file or URL

try:
    gdf = gpd.read_file(path)  # geopandas builds on pandas
except Exception as e:
    raise RuntimeError(f"Could not read '{path}': {e}")

# Reproject to Web Mercator for basemap tiles (if available)
gdf_3857 = gdf.to_crs(epsg=3857)

fig, ax = plt.subplots(figsize=(10, 10))
gdf_3857.plot(ax=ax, edgecolor="k", linewidth=0.5, alpha=0.7, cmap="viridis")
if has_ctx:
    ctx.add_basemap(ax, source=ctx.providers.Stamen.TonerLite)
ax.set_title("Pittsburgh Watersheds")
ax.set_axis_off()
plt.show()
```
