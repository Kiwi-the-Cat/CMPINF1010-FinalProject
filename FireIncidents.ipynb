```python
import pandas as pd

fireInc = pd.read_csv("dataSets/FireIncidents.csv")

#Normalize column names
fireInc.columns = fireInc.columns.str.strip().str.lower()

#Normalize neighborhood names
fireInc['neighborhood'] = fireInc['neighborhood'].astype(str).str.strip().str.lower()

#Group by neighborhood and count incidents
fire_counts = fireInc.groupby("neighborhood").size().reset_index(name='incident_count')

#Sort from least to greatest
fire_counts_sorted = fire_counts.sort_values(by='incident_count')

# Step 6: Widen display settings to show full output
pd.set_option("display.width", 200)
pd.set_option("display.max_rows", None)

print(fire_counts_sorted)
```
